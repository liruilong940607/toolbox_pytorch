```python
from cvlib.tools import get_affine_matrix, get_inverse_affine_matrix
```

```python
M = get_affine_matrix(center=[533/2.0, 400.0], angle=0, translate=[0,0], scale=[0.5, 1.5], shear=30)
M_inv = get_inverse_affine_matrix(center=[533/2.0, 400.0], angle=0, translate=[0,0], scale=[0.5, 1.5], shear=30)
```

```python
print M[:-1]
print M_inv[:-1]
```

```python
import cv2
img = cv2.imread('/home/dalong/testimg.png')
print img.shape

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 10)

plt.imshow(img)
plt.show()

img = cv2.warpAffine(img, M[:-1], (533, 800), flags=cv2.INTER_LINEAR)

plt.imshow(img)
plt.show()

img = cv2.warpAffine(img, M_inv[:-1], (533, 800), flags=cv2.INTER_LINEAR)

plt.imshow(img)
plt.show()
```
